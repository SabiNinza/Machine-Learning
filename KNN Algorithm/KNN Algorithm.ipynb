```python
import numpy as np
import matplotlib.pyplot as plt
```

# Generating Our DataSet

```python
#First normal Distribution
orange_mean=[8,10]
orange_cov=[[3,0.4],[0.4,5]]
orange=np.random.multivariate_normal(orange_mean,orange_cov,400)
#second normal distribution
apple_mean=[4,8]
apple_cov=[[5,-0.4],[-0.4,7]]
apple=np.random.multivariate_normal(apple_mean,apple_cov,400)
#third normal distribution
gauva_mean=[2,2]
gauva_cov=[[3,0.7],[0.7,3]]
gauva=np.random.multivariate_normal(gauva_mean,gauva_cov,200)

print(orange.shape,apple.shape,gauva.shape,sep=" ")
```

```python
#Now plot 1 ,2, 3 distribution on graph
plt.figure()
plt.scatter(orange[:,0],orange[:,1],color="orange",label="orange => 0")
plt.scatter(apple[:,0],apple[:,1],color="red",label="apple => 1")
plt.scatter(gauva[:,0],gauva[:,1],color="green",label="gauva =>2")
plt.legend()
plt.show()
```

```python
row=orange.shape[0]+apple.shape[0]+gauva.shape[0]
column=orange.shape[1]+1

dataset1=np.zeros((row,column))
dataset1.shape
```

```python
dataset1[ :orange.shape[0],:-1]=orange
s2=orange.shape[0]+apple.shape[0]
dataset1[orange.shape[0] : s2,:-1]=apple
dataset1[s2 : ,:-1]=gauva
#label
dataset1[:orange.shape[0],-1]=0
dataset1[orange.shape[0]:s2,-1]=1
dataset1[s2:,-1]=2
dataset1.shape
```

```python
np.random.shuffle(dataset1)#inplace
split1=int(0.8*dataset1.shape[0])
x1_train=dataset1[:split1,:-1]
x1_test=dataset1[split1:,:-1]

y1_train=dataset1[:split1,-1]
y1_test=dataset1[split1:,-1]

print(x1_train.shape,x1_test.shape,y1_train.shape,y1_test.shape)
```

# KNN Algorithm

### Knn Problem Statement

```python
point=np.array([2.5,5.0])
```

```python
plt.figure()
plt.scatter(orange[:,0],orange[:,1],color="orange",label="Orange => 0")
plt.scatter(apple[:,0],apple[:,1],color="red",label="apple => 1")
plt.scatter(gauva[:,0],gauva[:,1],color="green",label="gauva => 2")
plt.scatter(point[0],point[1],color="black",marker="*",s=500,label="Blank")
plt.legend()
plt.show()
```

### Implementing KNN

```python
def euclidean_Dis(l1,l2):
    return np.sqrt(np.sum((l1-l2)**2))
```

```python
euclidean_Dis(np.array([1,2,8]),np.array([4,7,6]))
```

```python
def KNN_Algo(x_train,y_train,test_point,k=5):
    distances=[]
    
    #compute the distace from testpoint
    for datapoint,label in zip(x_train,y_train):
        distance=euclidean_Dis(test_point,datapoint)
        distances.append((distance,label))
    
    #sort
    distances=sorted(distances,key=lambda x :x[0])
    #choose first k nearest neighbour
    k_near_neigh=np.array(distances[:k])
    freq=np.unique(k_near_neigh[:,1],return_counts=True)
    labels,counts=freq
    #print(k_near_neigh[:,1])
    #choose most freq label
    ans=labels[counts.argmax()]
    return ans
```

```python
KNN_Algo(x1_train,y1_train,point)
```

### Calculate the accuracy of "KNN Algorithm"

```python
def Get_Prediction(x_train,y_train,x_test,k=5):
    pred_x_test=[]
    
    for t_point in x_test:
        p=KNN_Algo(x_train,y_train,t_point,k)
        pred_x_test.append(p)
    return np.array(pred_x_test)    
```

```python
def Cal_Accuracy(pred_x_test,y_test,k=5):
    return (pred_x_test==y_test).sum()/y_test.shape[0]
```

```python
pred_x_test=Get_Prediction(x1_train,y1_train,x1_test)
#print(pred_x_test)
Cal_Accuracy(pred_x_test,y1_test)
```

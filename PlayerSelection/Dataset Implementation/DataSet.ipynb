# Generating Our DataSet

```python
#Example how to generate the Random Dataset with mean and covariance
mean=[5,6]
cov=[[3,-0.7],[-0.7,3]]
dist=np.random.multivariate_normal(mean,cov,600)

print(dist.shape)
#print(dist)
```

```python
plt.figure()
plt.scatter(dist[:,0],dist[:,1],color="purple")
plt.show()
```

## 2 distribution dataset

```python
#first Normal distribution
mean_01=[2,4]
cov_01=[[5,0.6],[0.6,6]]
dist_01=np.random.multivariate_normal(mean_01,cov_01,500)
#second normal distribution
mean_02=[10,8]
cov_02=[[8,-0.4],[-0.4,5]]
dist_02=np.random.multivariate_normal(mean_02,cov_02,500)

print(dist_01.shape)
print(dist_02.shape)
```

```python
#lets plot these two distribution point on graph
plt.figure()
plt.scatter(dist_01[:,0],dist_01[:,1],color="indigo")
plt.scatter(dist_02[:,0],dist_02[:,1],color="green")
#plot a point for which we want to know from which distribution it belong
plt.scatter(4,8,color="red",marker="*",s=500)
plt.show()
```

```python
#Now find row and column to build dataset of two distribution
#its obvious if there is two distribution then label also present
#which increase the one column
row=dist_01.shape[0]+dist_02.shape[0]
column=dist_01.shape[1]+1

dataset=np.zeros((row,column))  #it create array filled with 0
dataset.shape
```

```python
#Now we need to copy the datapoint of dist_01 or dist_02 to dataset
dataset[:dist_01.shape[0],:-1]=dist_01
dataset[dist_02.shape[0]:,:-1]=dist_02

#Now Assign a label for 2 distribution 
dataset[:dist_01.shape[0],-1]=0
dataset[dist_02.shape[0]:,-1]=1
```

```python
dataset[490:510]
```

```python
#From this dataset if we build our training and test dataset 
#there is large probabilty it contain 1 label or 0 label any one more in number 
#for avoiding this we shuffle our array so 0 and 1 are randomly shuffle
#so when we use  20percent for test dataset it contain 0 and 1 in equal probabilty
np.random.shuffle(dataset)
```

```python
dataset[490:510]
#dataset.shape
```

#### Spliting our dataset into training and testing dataset

```python
split=int(0.8*dataset.shape[0]) #80 % dataset for traning
x_train=dataset[:split,:-1]
x_test=dataset[split:,:-1]

y_train=dataset[:split,-1]
y_test=dataset[split:,-1]

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape,sep=" ")
```
